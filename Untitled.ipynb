```python
BASE_URL = "https://hn.algolia.com/api/v1/search_by_date"

def fetch_hn(keyword, tag="comment", hits_per_page=200, pages=2, sleep=0.3):
    """
    tag: 'comment' or 'story'
    pages: how many pages of results to fetch (each page up to hits_per_page)
    """
    all_hits = []
    for page in range(pages):
        params = {
            "query": keyword,
            "tags": tag,
            "hitsPerPage": hits_per_page,
            "page": page
        }
        r = requests.get(BASE_URL, params=params, timeout=30)
        r.raise_for_status()
        data = r.json()
        hits = data.get("hits", [])
        all_hits.extend(hits)
        time.sleep(sleep)
    return all_hits
```

```python
def hits_to_df(hits, keyword, content_type):
    rows = []
    for h in hits:
        # comments have 'comment_text', stories have 'title' + 'story_text'
        text = h.get("comment_text") or h.get("title") or h.get("story_text") or ""
        text = text.replace("\n", " ").strip() if isinstance(text, str) else ""
        
        rows.append({
            "source": "HackerNews",
            "keyword": keyword,
            "content_type": content_type,
            "objectID": h.get("objectID"),
            "created_at": h.get("created_at"),
            "author": h.get("author"),
            "points": h.get("points"),
            "num_comments": h.get("num_comments"),
            "url": h.get("url") or h.get("story_url"),
            "title": h.get("title"),
            "story_title": h.get("story_title"),
            "text": text
        })
    df = pd.DataFrame(rows)
    return df
```

```python
all_dfs = []

for kw in KEYWORDS:
    # Comments = opinions/feedback (usually most useful)
    comment_hits = fetch_hn(kw, tag="comment", pages=2)
    df_comments = hits_to_df(comment_hits, kw, "HN Comment")
    all_dfs.append(df_comments)

    # Stories = headlines/announcements
    story_hits = fetch_hn(kw, tag="story", pages=1)
    df_stories = hits_to_df(story_hits, kw, "HN Story")
    all_dfs.append(df_stories)

hn_df = pd.concat(all_dfs, ignore_index=True)

print("Total rows fetched:", len(hn_df))
hn_df.head(5)
```

```python
hn_df["text"] = hn_df["text"].fillna("").astype(str).str.strip()
hn_df = hn_df[hn_df["text"].str.len() > 20].copy()

hn_df = hn_df.drop_duplicates(subset=["objectID"]).reset_index(drop=True)

# parse timestamps
hn_df["created_at"] = pd.to_datetime(hn_df["created_at"], errors="coerce", utc=True)
hn_df["fetched_at"] = datetime.now(timezone.utc)

hn_df.shape
```

```python
def clean_hn_text(text):
    if not isinstance(text, str):
        return ""
    
    # Decode HTML entities (&gt;, &quot;, etc.)
    text = html.unescape(text)
    
    # Remove HTML tags using BeautifulSoup
    soup = BeautifulSoup(text, "html.parser")
    text = soup.get_text(separator=" ")
    
    # Remove URLs
    text = re.sub(r"http\S+|www\S+", "", text)
    
    # Remove code-like patterns (long inline code)
    text = re.sub(r"\b[A-Za-z0-9_]{15,}\b", "", text)
    
    # Remove extra whitespace
    text = re.sub(r"\s+", " ", text).strip()
    
    return text
```

```python


def clean_hn_text(t: str) -> str:
    if pd.isna(t):
        return ""
    t = html.unescape(str(t))                 # converts &gt; &quot; &#x27; etc.
    t = re.sub(r"<[^>]+>", " ", t)            # remove HTML tags like <p>, <i>, <a>
    t = re.sub(r"\s+", " ", t).strip()        # normalize spaces
    return t

df["clean_text"] = df["text"].apply(clean_hn_text)
df["clean_length"] = df["clean_text"].str.len()

df[["text","clean_text","clean_length"]].head(3)
```

```python


TAG_RE = re.compile(r"<[^>]+>")
URL_RE = re.compile(r"https?://\S+|www\.\S+")
CODEBLOCK_RE = re.compile(r"(?is)<pre><code>.*?</code></pre>")  # remove code blocks

def multi_unescape(s: str, n=3) -> str:
    # Some HN text is double-encoded, so unescape a few times
    for _ in range(n):
        new = html.unescape(s)
        if new == s:
            break
        s = new
    return s

def clean_hn_text_strong(t: str) -> str:
    if pd.isna(t):
        return ""

    s = str(t)

    # 1) Remove code blocks early (before stripping tags)
    s = CODEBLOCK_RE.sub(" ", s)

    # 2) Decode HTML entities (often multiple times)
    s = multi_unescape(s, n=5)

    # 3) Remove remaining HTML tags
    s = TAG_RE.sub(" ", s)

    # 4) Remove URLs
    s = URL_RE.sub(" ", s)

    # 5) Remove HN-style quoting markers / leftover arrows
    s = re.sub(r"(^|\s)>\s*", " ", s)

    # 6) Kill leftover HTML-ish artifacts and normalize
    s = s.replace("\u00a0", " ")
    s = re.sub(r"\s+", " ", s).strip()

    return s

df["clean_text"] = df["text"].apply(clean_hn_text_strong)
df["clean_length"] = df["clean_text"].str.len()

df[["text","clean_text"]].head(1)
```
